```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = pd.read_csv('dog_rates_tweets.csv', parse_dates=True)
```

```python
#data
```

```python
extract_data = data.text.str.extract(pat = r'(\d+(\.\d+)?)/10')
```

```python
extract_data[0] = pd.to_numeric(extract_data[0])
```

```python
filter_data = extract_data[extract_data[0] <= 25]
```

```python
#filter_data
```

```python
graph_data = data.loc[filter_data.index]
graph_data['rating'] = filter_data[0]
```

```python
def to_timestamp(x):
    return x.timestamp()
```

```python
graph_data['created_at'] = pd.to_datetime(graph_data['created_at'], format='%Y-%m-%d %H:%M:%S')
graph_data['timestamp'] = graph_data['created_at'].apply(to_timestamp)
```

```python
fit = stats.linregress(graph_data['timestamp'], graph_data['rating'])
```

```python
fit.slope, fit.intercept
```

```python
graph_data['prediction'] = graph_data['timestamp']*fit.slope + fit.intercept 
```

```python
graph_data
```

```python
import seaborn
seaborn.set()

plt.xticks(rotation=25)
plt.plot(graph_data['created_at'], graph_data['rating'], 'b.', alpha=0.5)
plt.plot(graph_data['created_at'], graph_data['prediction'], 'r-', linewidth=3)
plt.title('Month Vs Dog Ratings')
plt.xlabel('Months')
plt.ylabel('Ratings')
```

```python
#p-value of regression line
fit.pvalue
```

```python
hist_data = pd.Series(graph_data['rating'] - graph_data['prediction'])
```

```python
#hist_data
```

```python
plt.hist(hist_data)
plt.show()
```
